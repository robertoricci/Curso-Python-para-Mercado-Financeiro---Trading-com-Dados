# Agora vamos construir uma carteira e comparar com o IBOV

```python
carteira = ['PETR4','VALE3', 'ABEV3', 'WEGE3', 'EMBR3', 'RADL3', 'CVCB3', 'CIEL3', 'OIBR3', 'KNRI11']
```

```python
def retorna_ativos(ativos, inicio, fim):
  df = pd.DataFrame()

  for i in ativos: 
    ticker = str(i) + '.SA'
    df[i] = yf.download(ticker, start = inicio, end = fim)['Adj Close']

  return df
```

```python
inicio = '2020-01-01'
fim = '2022-07-12'
precos = retorna_ativos(carteira, inicio, fim)
```

```python
precos.head()
```

### Quantos papéis compramos de cada um desses ativos?

```python
#pegar primeira linha
primeira = precos.iloc[0]
```

```python
primeira
```

```python
# temos 10000, vamos dividir 1000 em cada ativo
valor_inv = 10000
comprados = round((valor_inv/10)/primeira, 0)
```

```python
# verficar os valores dos papeis.
valores = comprados * primeira
```

```python
valores
```

#Calcular o Patrimonio liquido ao longo do tempo, suponte que não foi vendido nenhum papel.

```python
pl = precos*comprados
```

```python
pl
```

```python
#criar um coluna para somar tudo 
                       #fazer em todas linhas somar no sentido da coluna
pl['Minha carteria'] = pl.iloc[:].sum(axis= 1 )
```

```python
pl.tail()
```

```python
pl['Minha carteria'].plot();
```

# Baixando dados do IBOV

```python
ibov = yf.download('^BVSP',start= inicio,end=fim)
```

```python
ibov.rename(columns = {'Adj Close': 'IBOV'},inplace=True)
```

```python
ibov
```

```python
#deixar somente a coluna IBOV
ibov = ibov.drop(ibov.columns[[0,1,2,3,5]],axis=1)
```

```python
ibov
```

```python
#garantir que os index da
pl.index = pd.to_datetime(pl.index)
```

```python
#garantir que os index da
ibov.index = pd.to_datetime(ibov.index)
```

```python
novo_df = pd.merge(ibov, pl, how = 'inner', on = 'Date')
```

```python
novo_df.tail()
```

```python
normalizado = novo_df/novo_df.iloc[0]
```

```python
normalizado.head()
```

```python
#trazendo dados do ibov e da carteira da ultima data
normalizado.iloc[-1,[0,11]]
```

```python
normalizado[['IBOV','Minha carteria']].plot();
```

```python
perclucro =( normalizado.iloc[-1,11] - 1 ) * 100
vlr = perclucro/100 * valor_inv

print("podemos ver que a carteira montada bateu o IBOV, porém devido algumas quedas recentes, nossa carteira tem um lucro/prejuizo de {} ".format(perclucro))
print("ou seja se precisar vender as ações hoje o prejuizo em valor e {}".format(vlr)) 
```
